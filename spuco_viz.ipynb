```python
baseline_log_f = "./spuco_results/baseline/baseline_log.json"
jtt_log_f = "./spuco_results/jtt/jtt_log.json"
spare_log_f = "./spuco_results/spare/spare_log.json"
baseline_best_eval_f = "./spuco_results/baseline/baseline_best_evalution.json"
jtt_best_eval_f = "./spuco_results/jtt/jtt_best_evalution.json"
spare_best_eval_f = "./spuco_results/spare/spare_best_evalution.json"
baseline_last_eval_f = "./spuco_results/baseline/baseline_last_evalution.json"
jtt_last_eval_f = "./spuco_results/jtt/jtt_last_evalution.json"
spare_last_eval_f = "./spuco_results/spare/spare_last_evalution.json"
```

```python
import json

with open(baseline_log_f) as f1, \
     open(jtt_log_f) as f2,\
     open(spare_log_f) as f3:
    baseline_log = json.load(f1)["0"]["losses"]
    jtt_log = json.load(f2)["0"]["losses"]
    spare_log = json.load(f3)["0"]["losses"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

ind = np.arange(len(baseline_log)) # 1581
jtt_log = jtt_log[0:1581]

baseline_mov_avg = []

plt.figure(figsize=(12,4))
plt.plot(ind, jtt_log, color='g', label="jtt loss", alpha=0.65)
plt.plot(ind, spare_log, color='b', label="spare loss", alpha=0.65)
plt.plot(ind, baseline_log, color='r', label="baseline loss")
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
with open(baseline_best_eval_f) as f1, \
     open(baseline_last_eval_f) as f2, \
     open(jtt_best_eval_f) as f3, \
     open(jtt_last_eval_f) as f4, \
     open(spare_best_eval_f) as f5, \
     open(spare_last_eval_f) as f6:
    baseline_eval = json.load(f1)
    baseline_group_acc, baseline_worst_acc, baseline_avg_acc, baseline_best_acc = baseline_eval["group_accuracy"], baseline_eval["worst_accuracy"], baseline_eval["avg_accuracy"], baseline_eval["best_accuracy"]
    # baseline_last_eval = json.load(f2)
    jtt_eval = json.load(f3)
    jtt_group_acc, jtt_worst_acc, jtt_avg_acc, jtt_best_acc = jtt_eval["group_accuracy"], jtt_eval["worst_accuracy"], jtt_eval["avg_accuracy"], jtt_eval["best_accuracy"]
    # jtt_last_eval = json.load(f4)
    spare_eval = json.load(f5)
    spare_group_acc, spare_worst_acc, spare_avg_acc, spare_best_acc = spare_eval["group_accuracy"], spare_eval["worst_accuracy"], spare_eval["avg_accuracy"], spare_eval["best_accuracy"]
    # spare_last_eval = json.load(f6)
```

```python
ind = np.arange(len(baseline_group_acc))
plt.figure(figsize=(18,6))
plt.plot(ind, list(baseline_group_acc.values()), 'o--', color='r', linewidth=1, label="baseline group acc")
plt.plot(ind, list(jtt_group_acc.values()), 'v--', color='g', linewidth=1, label="jtt group acc")
plt.plot(ind, list(spare_group_acc.values()), 's--', color='b', linewidth=1, label="spare group acc")
plt.xticks(ind, list(baseline_group_acc.keys()), fontsize=6)
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
baseline_acc = [baseline_worst_acc, baseline_avg_acc, baseline_best_acc]
jtt_acc = [jtt_worst_acc, jtt_avg_acc, jtt_best_acc]
spare_acc = [spare_worst_acc, spare_avg_acc, spare_best_acc]

ind = np.arange(len(baseline_acc))
plt.plot(ind, baseline_acc, 'o--', color='r', linewidth=1, label="baseline")
plt.plot(ind, jtt_acc, 'v--', color='g', linewidth=1, label="jtt")
plt.plot(ind, spare_acc, 's--', color='b', linewidth=1, label="spare")
plt.xticks(ind, ["worst acc", "average acc", "best acc"])
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
